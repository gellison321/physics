# Exercise 1
Given a force that varies with time according to $F=2t^{2}$, and with the initial condition at time zero, $x(0)= \pi$, use Aristotle's law to find $x(t)$ at all times.

Aristotle's law is given by:

$\vec{F} = m\vec{v}$

$x(t)$, position, is the time integral of velocity, so $2t^{2} = m\cdot \dot{\vec{x}} $

$\implies \dot{\vec{x}} = \frac{2t^{2}}{m}$

$\implies \vec{v} = \int \dot{\vec{x}} dt = \int \frac{2t^{2}}{m} dt$


$\frac{1}{m}\int2t^{2} dt = \frac{2}{3m}t^{3} + c$

$\pi = \frac{2}{3m}0^{3} + c \implies c = \pi$

$\implies x(t) = \frac{2}{3m}t^{3} + \pi$

Show by differentiation that the general solution to 

$\ddot{x} = -w^{2}x$ 

is given in terms of two constants, A and B, by the position equation

$x(t) = Acos(wt) + sin(wt)$

Determine the initial position and velocity at time $t= 0$ in terms of A and B

Start by differentiating the function x(t)

$\dot{x(t)} = wBcos(wt) - wAsin(wt)$

$\ddot{x(t)} = -w^{2}Bsin(wt) - w^{2}Acos(wt)$

Plugging into the differential equation

$\implies -w^{2}Bsin(wt) - w^{2}Acos(wt) = -w^{2}[Acost(wt) + Bsin(wt)]$

Finding the initial position in terms of A and B

$x(0) = Acos(0w) + Bsin(0w) = A$

Finding the initial velocity in terms of A and B

$\dot{x(0)} = wBcos(w0) - wAsin(w0) = wB$
