# Exercise 1: 
Plot the following functions

$f(t) = t^{4}+3t^{3}-12t^{2}+t-6$

$g(x) = sin(x) - cos(x)$

$\theta(\alpha) = e^{\alpha}+\alpha \ln(\alpha)$

$x(t) = sin^{2}(t) - cos(t)$

```python
def f(t): return t**4 + 3*t**3 - 12*t**2 + t - 6

arange = np.arange(-10, 10, .1)

plt.plot(arange, [f(t) for t in arange])
plt.xticks(np.arange(min(arange), max(arange)+1, 1.0));
```

```python
def g(x): return np.sin(x) - np.cos(x)

arange = np.arange(-10, 10, .1)

plt.plot(arange, [g(x) for x in arange])
plt.xticks(np.arange(min(arange), max(arange)+1, 1.0));
```

```python
def theta(alpha): return np.exp(alpha) + alpha*np.log(alpha)

arange = np.arange(1, 10, .1)

plt.plot(arange, [theta(alpha) for alpha in arange])
plt.xticks(np.arange(min(arange), max(arange)+1, 1.0));
```

```python
def x(t): return np.sin(t)**2 - np.cos(t)

arange = np.arange(-10, 10, .1)

plt.plot(arange, [x(t) for t in arange])
plt.xticks(np.arange(min(arange), max(arange)+1, 1.0));
```

# Exercise 2
Work out the rule for vector subtraction

$\vec{a} - \vec{b} = \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{bmatrix} - \begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix} = \begin{bmatrix} a_1 - b_1 \\ a_2 - b_2 \\ \vdots \\ a_n - b_n \end{bmatrix}
$

```python
a = [np.random.randint(20) for _ in range(10)]
b = [np.random.randint(20) for _ in range(10)]

# solution using list comprehension
c = [i-j for i,j in zip(a,b)]
print(c)

# or solution using numpy
c = np.array(a) - np.array(b)
print(c)
```

# Exercise 3
Show that the magnitude of the vector satisfies

$|\vec{A}|^{2} = \vec{A} \cdot \vec{A}$

The magnitude of a vector is given by: $|\vec{A}| = \sqrt{A_{0}^{2} + A_{1}^{2} + ... +  A_{n}^{2}}$

$\iff|\vec{A}|^{2} = A_{0}^{2} + A_{1}^{2} + \dots +  A_{n}^{2} = \sum_{i=1}^{n}A_{i}^{2}$

The dot product of a vector is given by:

$\vec{A} \cdot \vec{B} = A_{1}B_{1} + A_{2}B_{2} + \dots + A_{n}B_{n} = \sum_{i=1}^{n} A_{i}B_{i}$

$\iff |\vec{A}|^{2} = \vec{A} \cdot \vec{A}$



```python
A = [np.random.randint(20) for _ in range(10)]

magnitude = np.sqrt(sum([a**2 for a in A]))

dot_product = sum([a*a for a in A])

assert np.allclose(magnitude**2, dot_product, atol=1e-10)
```

# Exercise 4
Let $(A_{x} = 2, A_{y} = -3, A_{z} = 1)$

and $(B_{x} = -4, B_{y} = -3, B_{z} = 2)$

Compute the magnitude of $\vec{A}$ and $\vec{B}$, their dot product, and the angle between them


$|\vec{A}| = \sqrt{2^{2} + -3^{2} + 1^{2}} = 3.7417$

$|\vec{B}| = \sqrt{-4^{2} + -3^{2} + 2^{2}} = 5.3852$

$\vec{A} \cdot \vec{B} = 2\cdot(-4) + (-3)\cdot(-3) + 1\cdot2 = 3$

The relationship: $cos(\theta) = \frac{\vec{A}\cdot\vec{B}}{|\vec{A}||\vec{B}|}$

$cos(\theta) = \frac{3}{3.7417 \cdot 5.3852} = 4.3177$

$\iff \theta = cos^{-1}(\frac{\vec{A}\cdot\vec{B}}{|\vec{A}||\vec{B}|}) = 1.4214$

```python
A = [2, -3, 1]
B = [-4, -3, 2]

magnitude_A = np.linalg.norm(A)
magnitude_B = np.linalg.norm(B)

dot_A_B = np.dot(A, B)

cos_angle = dot_A_B/(magnitude_A*magnitude_B)

angle_radians = np.arccos(cos_angle)

angle_degrees = np.degrees(angle_radians)
```
